```python
import json
from tqdm.auto import tqdm

line_count = len(open("../data/photo.json").readlines())
caption, photo_id, business_id, label = [], [], [], []
with open("../data/photo.json") as f:
    for line in tqdm(f, total=line_count):
        blob = json.loads(line)
        caption += [blob["caption"]]
        photo_id += [blob["photo_id"]]
        business_id += [blob["business_id"]]
        label += [blob["label"]]

photo_df = pd.DataFrame(
 {"caption": caption, "business_id": business_id, "label": label, "photo_id": photo_id}
)
```
